```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
```

```python
barchart = titanic[(titanic["pclass"] == 1) & (titanic["embarked"] == "S")]
```

```python
plt.title('First Class Titanic Passengers that left from Southampton'+ '\n'+ 'grouped by survival and gender', fontsize = 20)
sns.countplot(x="survived", hue="sex", color="g", data = titanic)
plt.xticks([0,1],["No","Yes"])
plt.xlabel('Survived')
```

```python
gdp = pd.read_excel("GDP.xlsx")
```

```python
gdp = gdp.drop(gdp.index[0:10])
```

```python
fig, ax = plt.subplots()

plt.title('U.S. GDP since 1990')
plt.xlabel('Year')
plt.ylabel('GDP ($billions)')

plt.plot(gdp["Year"], gdp["GDP"])

ax.annotate('Start of Global'+'\n'+'Financial Crisis', xy=(2008, 14500), xytext=(2004, 16000),
            arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
baseball = pd.read_excel("Baseball.xlsx")
```

```python
plt.scatter(baseball["Runs Scored per Game"], baseball["Wins"])
plt.xlabel("Runs Scored per Game")
plt.ylabel("Wins")
plt.title("Runs Scored per Game compared to Wins in MLB")
plt.text(3.77, 92, 'St. Louis', color='red', fontsize=12)
plt.text(4.61, 68, 'Colorado ', color='purple', fontsize=12)
plt.text(3.25, 94, 'On average, the more runs\n a team scores, the more\n games the team will win', color='grey', fontsize=8)
```
