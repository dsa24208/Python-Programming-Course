# Hard for SPACs

```python
import pandas as pd
import smtplib
from yahoo_fin import stock_info as si
```

```python
#steals all spac data from some schmuck
spacs = pd.read_html("https://stockmarketmba.com/listofshellcompanies.php")
```

```python
#idk why this works to move it from a list to a df, but oh well
spac = spacs[0]
```

```python
#keeps relevant columns
spacdata = spac[["Symbol", "Name", "IPO date", "Merger pending?"]].copy()
```

```python
#drops null values
spacdata.dropna(inplace = True)
```

```python
spaclist = spacdata.Symbol
```

```python
spaclist.head(60)
```

```python
stock_data = pd.DataFrame(columns = ['Ticker', 'Market Cap', 'Open', 'Close', 'Volume'])
```

```python
for i in spaclist:
    try:
        data = si.get_data(i).tail(1)
        quote_table = si.get_quote_table(i)
        stock_data = stock_data.append({'Ticker' : data.iloc[0]['ticker'], 'Market Cap' : quote_table['Market Cap'], 
                                    'Open' : data.iloc[0]['open'], 'Close' : data.iloc[0]['close'], 
                                    'Volume' : data.iloc[0]['volume']}, ignore_index = True)
    except:
        pass
```

```python
spacwatch = pd.merge(spacdata, stock_data, left_on = 'Symbol', right_on = 'Ticker')
spacwatch.drop('Ticker', axis = 1, inplace = True)
```

```python
#sorts options for SPACs to be used in final
cheapspacsnomerge = spacwatch.sort_values([ "Merger pending?", "Close"], ascending = (True, True)).head(100).reset_index(drop = True)
cheapspacswithmerge = spacwatch.sort_values([ "Merger pending?", "Close"], ascending = (False, True)).head(50).reset_index(drop = True)
popularspacs = spacwatch.sort_values("Volume", ascending = False).head(50).reset_index(drop = True)
```

```python
#creates and sends out email with correct formatting

from smtplib import SMTP
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

def send_mail(recipient):

    message = MIMEMultipart()
    message['Subject'] = 'SPAC/INSIDER UPDATE'
    message['From'] = 'dolceandgarbage@gmail.com'
    message['To'] = recipient

    body_content = """\
    <html>
      <head></head>
      <body>
        <p><b>Let's go to the <sup>moon.</sup></b><br>
           <br>
           <b>Cheap premerger SPACS:</b><br>
           {0}
           <br>
           <br>
           <b>Highest Volume SPACs:</b><br>
           <br>
           {1}
           <br>
           <br>
           <b>Cheap postmerger SPACs:</b><br>
           <br>
           {2}
           <br>
           <br>
           <b>Lastest SEC SPAC filings - for more info visit https://sec.report/SPAC.php:</b><br>
           <br>
           {3}
           <br>
           <br>
           <b>Most recent insider buying:</b><br>
           <br>
           {4}
           
        </p>
      </body>
    </html>

    """.format(cheapspacsnomerge.to_html(), popularspacs.to_html(), cheapspacswithmerge.to_html(), newspacs.to_html(), recent_insiders.to_html())
    message.attach(MIMEText(body_content, "html"))
    msg_body = message.as_string()

    server = SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(message['From'], 'Sund3vil')
    server.sendmail(message['From'], message['To'], msg_body)
    server.quit()
    
```
